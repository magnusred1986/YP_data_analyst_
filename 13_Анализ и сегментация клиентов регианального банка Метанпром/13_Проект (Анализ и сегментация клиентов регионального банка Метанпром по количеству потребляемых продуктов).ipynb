Начнем с наименования столбцов, приведем в соответсвие  

```python
#напишем функцию перевода столбцов к нижнему регистру и переведем в "питонячий формат))"  "_" 
# а следом цикл, который применит функцию

list_df = [df] # список с df для изменений

def replace_columns_names(df):
    list_of_columns = []
    for column in df.columns:
        column = column.lower()
        column = column.replace(' ', '_')
        list_of_columns.append(column)
    df.columns = list_of_columns
    return df

# применяем функцию к нашим df
for i in list_df:               # пробегаем по списку с df и подаем в обработку
    replace_columns_names(i)    # применение функции
    display(list(i.columns))    # вывод результата для визуальной оценки изменений
```

### Проверка связей по `шкале Чеддока`    

<a name="stage_1"></a>

[Наверх к оглавлению](#contents)

**Шкала Чеддока:**   
η : степень тесноты связи  
0,1–0,3 слабая  
0,3–0,5 умеренная  
0,5–0,7 заметная  
0,7–0,9 высокая  
0,9–0,99 сильная  
0,99–1 функциональная  

Напишем функцию  

```python
def corr_info(x):
    if x <= 0.3:
        return 'корреляция слабая'
    if 0.3 < x <= 0.5:
        return 'корреляция умеренная'
    if 0.5 < x <= 0.7:
        return 'корреляция заметная'
    if 0.7 < x <= 0.9:
        return 'корреляция высокая'
    if 0.9 < x <= 0.99:
        return 'корреляция сильная'
    if x > 0.99:
        return 'корреляция очень высокая'
```

```python
test_x = 'age' # параметр который тестируем к остальным значениям
spisok_ignor_corr = ['city', 'gender'] # столбцы игнорирования где не числовые значения
for i in df.columns:
    if i not in spisok_ignor_corr and i != test_x: # исключаем список игнора и само тестируемое значение
        print(test_x,'к',i, df[test_x].corr(df[i]).round(2), corr_info(df[test_x].corr(df[i])))
```

* согласно `Шкале Чеддока`, корреляция везде слабая

## Проверка гипотез

Проверим гипотезы  

Критерий t Стьюдента направлен на оценку различий величин средних $\overline
Х $ и $\overline У $ двух выборок X и Y, которые распределены по нормальному закону.   

Для применения t-критерия Стьюдента необходимо, чтобы исходные данные имели нормальное распределение. Также имеет значение равенство дисперсий (распределения) сравниваемых групп. При неравных дисперсиях применяется t-критерий в модификации Уэлча.

### `Гипотеза` различия дохода между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним    

<a name="stage_1"></a>

[Наверх к оглавлению](#contents)

```python
df_products_2=df.query("products==2")
df_products_1=df.query("products==1")
```

```python
df.query('products <= 2').groupby('products')['estimated_salary'].agg(['mean', 'median', 'std', 'count'])
```

```python
plt.figure(figsize=(18, 8)) # размер фигуры
sns.distplot(df_products_1['estimated_salary'],label='1 продукт')
sns.distplot(df_products_2['estimated_salary'],label='2 продукта')
plt.title('Распределение estimated_salary - заработной платы клиента',fontsize=16, fontweight = 'bold')
plt.legend()
plt.xticks(rotation=0)
plt.show()
```

Так как все признаки, по которым мы будем проводить проверку гипотез, имеют распределение, отличное от нормального, будем использовать для проверки непараметрический критерий `Манна-Уитни`. Уровень статистической значимости принимаем равным `0,05`.

Проверим `гипотезу` различия дохода между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним  
`H0`- средняя выручка (заработная плата) между клиентами пользующимия 2 и 1 продуктом банка `равна`  
`H1` - средняя выручка (заработная плата) между клиентами пользующимия 2 и 1 продуктом банка `отличается`    

<div class="alert alert-success"> 
    
Гипотезы сформулированы верно
    
<div>

```python
print('Дисперсия клиентов пользующихся 2-мя продуктами:',df_products_2['estimated_salary'].var())
print('Дисперсия клиентов пользующихся 1-им продуктом:',df_products_1['estimated_salary'].var())
```

```python
# задаём уровень критерия статистической значимости
alpha=0.05
# формируем выборки 
# equal_var=False - равны ли дисперсии двух выборок (в нашем случае - нет)
results = st.ttest_ind(df_products_1['estimated_salary'], df_products_2['estimated_salary'], equal_var=False)
# получаем значение p-value
print('p-значение:', results.pvalue)

mean_1 = df_products_1['estimated_salary'].mean()
mean_2 = df_products_2['estimated_salary'].mean()
difference_ = abs(mean_1-mean_2)

# сравниваем полученное значение с критерием
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("\033[1m" +"Не получилось отвергнуть нулевую гипотезу"+"\033[0;0m")
    
print('Среднее по продукту 1 ', mean_1.round(2))
print('Среднее по продукту 2 ', mean_2.round(2))
print('Разница средних ', difference_.round(2))
```

Гипотезу `H0`- `отвергнуть не удалось` - средняя выручка (заработная плата) между клиентами пользующимия 2 и 1 продуктом банка `равна`  

### `Гипотеза` различия дохода между теми клиентами, которые ушли и теми, которые остались   

<a name="stage_1"></a>

[Наверх к оглавлению](#contents)

Проверим `гипотезу` различия дохода между теми клиентами, которые ушли и теми, которые остались  
`H0`- средняя выручка (заработная плата) между оставшимися и ушедними пользователями `равна`  
`H1` - средняя выручка (заработная плата) между оставшимися и ушедними пользователями `отличается`  

```python
df_churn_0=df.query("churn==0")
df_churn_1=df.query("churn==1")
```

```python
df.groupby('churn')['estimated_salary'].agg(['mean', 'median', 'std', 'count'])
```

```python
plt.figure(figsize=(18, 8)) # размер фигуры
sns.distplot(df_churn_1['estimated_salary'],label='1 - ушли ')
sns.distplot(df_churn_0['estimated_salary'],label='0 - остались ')
plt.title('Распределение estimated_salary - заработной платы клиента',fontsize=16, fontweight = 'bold')
plt.legend()
plt.show()
```

```python
print('Дисперсия ушедних:',df_churn_1['estimated_salary'].var())
print('Дисперсия оставшихся:',df_churn_0['estimated_salary'].var())
```

```python
# задаём уровень критерия статистической значимости
alpha=0.05
# формируем выборки
# equal_var=False - равны ли дисперсии двух выборок (в нашем случае - нет)
results = st.ttest_ind(df_churn_0['estimated_salary'], df_churn_1['estimated_salary'], equal_var=False)
# получаем значение p-value
print('p-значение:', results.pvalue)

mean_0 = df_churn_0['estimated_salary'].mean()
mean_1 = df_churn_1['estimated_salary'].mean()
difference = abs(mean_0-mean_1)

# сравниваем полученное значение с критерием
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("\033[1m" + "Не получилось отвергнуть нулевую гипотезу" + "\033[0;0m")
print('Среднее по ушедшим ', mean_0.round(2))
print('Среднее по оставшимся ', mean_1.round(2))
print('Разница средних ', difference.round(2))
```

Гипотезу `H0`- `отвергнуть не удалось` - средняя выручка (заработная плата) между оставшимися и ушедними пользователями `равна`.

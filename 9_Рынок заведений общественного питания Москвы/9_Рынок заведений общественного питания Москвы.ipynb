### Для какого вида объекта общественного питания характерно сетевое распространение?

<a name="stage_1"></a>

[Наверх к оглавлению](#contents)

Соберем общее количество заведений

```python
all_object = df.groupby('object_type')['object_name'].count().sort_values()
all_object
```

Соберем общее количество `сетевых` заведений

```python
net = df.query('chain == "сетевое"').groupby('object_type')['object_name'].count().sort_values()
net
```

Соберем общее количество `несетевых` заведений

```python
not_net = df.query('chain == "несетевое"').groupby('object_type')['object_name'].count().sort_values()
not_net
```

Объединим данные в один df

```python
sborka = pd.concat([all_object, net, not_net], axis = 1, sort=True).reset_index() # обьединил данные
```

Переименуем столбцы

```python
sborka.columns=['Наименование','всего','сетевые','несетевые'] # переименовали столбцы
```

Добавим сотлбцы с % долями 

```python
sborka['%_сетевых'] = (sborka['сетевые']/sborka['всего']*100).round(1)
sborka['%_несетевых'] = (sborka['несетевые']/sborka['всего']*100).round(1)
```

Взглянем на данные

```python
sborka
```

Отсортируем по убыванию по %_сетевых

#### Распределение в % общие данные 

```python
sborka.sort_values('%_сетевых', ascending = False)
```

Выведем ТОП-4

```python
sborka.sort_values('%_сетевых', ascending = False).head(4)
```

Таким образом, выявлены лидеры в сетевом сегменте

#### Визуализируем это в виде графиков:

График распределения доли сетевых/несетевых объектов среди заведений общественного питания

```python
with plt.style.context('seaborn'): # передали графику стиль
    plt.figure(figsize=(18, 8)) # зададим размер
    sns.stripplot(x='object_type', y='id', data=df, hue = 'chain')
    plt.title('Распределение сетевых/несетевых объектов среди заведений общественного питания',
              fontsize = 15, fontweight = 'bold') # заголовок 
    plt.legend(bbox_to_anchor=(0.01,1.1), framealpha=0.25, prop={'size':10}) # расположение и размер легенд
    plt.xlabel('Тип объекта',fontsize = 15, fontweight="bold") # ось Х
    plt.ylabel('кол-во',fontsize = 15, fontweight="bold") # ось Y
```

График распределения по количеству

```python
with plt.style.context('seaborn'): # стиль 
    sns.catplot(x='object_type', hue='chain', kind='count', data=df, height=8, aspect=6/4) # размеры height=8, aspect=8/4
    plt.title('Распределение количества сетевых/несетевых объектов среди заведений общественного питания',
              fontsize = 15, fontweight = 'bold') # заголовок 
    plt.xlabel('Тип объекта',fontsize = 15, fontweight="bold") # ось Х
    plt.ylabel('кол-во',fontsize = 15, fontweight="bold") # ось Y
```

Здесь наглядно видно, что сетевое распространение характерно для `кафе, фастфудов, ресторана и кулинарии`

```python
# сохранили в переменную список уникальных объектов
object_pit = list(df['object_type'].unique())
```

Напишем цикл и взглянем на каждый тип объекта отдельно

```python
with plt.style.context('seaborn'): # передали графику стиль
    for i in object_pit:
        df.query('object_type == @i').groupby(by='chain').agg({'chain':'count'}).plot(
            kind='pie',subplots=True, autopct='%1.1f%%', figsize = (18,10),fontsize=15,shadow=True)
        plt.legend(bbox_to_anchor=(0.01,1.1), framealpha=0.25, prop={'size':10})
        plt.title('Распределение на сетевое/несетевое по обьекту питания' + " - " +i, fontsize = 15, fontweight = 'bold')
```

Таким образом выяснили, что `сетевое распространение` характерно для следующих типов:  
* `фастфуд    41.1 %`  
* `кулинария  28.6 %`

### Посмотрите на распределение количества посадочных мест для улиц с большим количеством объектов общественного питания. Какие закономерности можно выявить?

<a name="stage_1"></a>

[Наверх к оглавлению](#contents)

```python
with plt.style.context('bmh'): # стиль     
    plt.figure(figsize=(18, 8)) # размер фигуры
    temp = df[df['street_name'].isin(top_list)]
    ax = sns.boxplot(x=temp['number'])
    ax.set_xlim(-10, persentil) # persentil - переенная со значением п.2.4.
    plt.title('Распределение количества посадочных мест для улиц с большим количеством объектов общественного питания',
              fontsize = 15, fontweight = 'bold') # заголовок 
    plt.show()
```

```python
with plt.style.context('bmh'): # стиль    
    plt.figure(figsize=(18, 8)) # размер фигуры
    ax = sns.boxplot(x="street_name", y="number", data=temp, orient='v')
    plt.title('Распределение количества посадочных мест для топ-10 улиц по кол-ву объектов',
              fontsize = 15, fontweight = 'bold') # заголовок 
    plt.xticks(rotation=45)
    ax.set_ylim(-10,persentil)
    plt.show()
```

```python
print('Среднее кол-во посадочных мест в заведениях на топ-10 улицах по количеству объектов:', temp['number'].median())
print('Среднее кол-во посадочных мест в заведениях НЕ на топ-10 улицах по количеству объектов:', df[-df['street_name'].isin(top_list)]['number'].median())


print('\nПроверим, значима ли статистически разница между кол-вами посадочных мест'
'\nв заведениях на топ-10 улицах и остальных.')
print('Используем тест Стъюдента.')
print('Нулевой гипотезой будет гипотеза о равенстве количества посадочных мест на топ-10 улицах по'
'\nколичеству объектов и на остальных улицах.')

alpha=0.05

results = stats.ttest_ind(df[-df['street_name'].isin(top_list)]['number'],
                            temp['number'])
print('P-value: {}'.format(results.pvalue))
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, разница между количествами посадочных мест статистически ЗНАЧИМА.')
else:
    print('Не отвергаем нулевую гипотезу. Разницы между количествами посадочных мест НЕТ.')
```

```python
df.columns.tolist() # взгялнем на названия
```

```python
# приведем наименование некоторых столбцов в приемелемый вид (переименование)
df = df.rename(columns={'cityCenters_nearest' : 'city_centers_nearest', 
                       'parks_around3000' : 'parks_around_3km',
                       'ponds_around3000' : 'ponds_around_3km'})
```

```python
df.duplicated().sum() # проверим на дубликаты, их нет
```

```python
df.columns # выведем наименование столбцов, будем обрабатывать только необходимые для расчетов
```

```python
#days_exposition - сколько дней было размещено объявление (от публикации до снятия)
# заменим пропуски и изменим тип данных
df['days_exposition'] = df['days_exposition'].fillna(0).astype('int')
```

```python
# посмотрим locality_name — название населённого пункта
# пустых значений 49, значения без локации лучше удалить, да и кол-во строк не велико
df['locality_name'].isnull().value_counts()
```

```python
# удалим строки БЕЗ ЛОКАЦИИ из df
df = df.dropna(subset=['locality_name'])
```

```python
# проверим
df['locality_name'].isnull().value_counts()
```

```python
# first_day_exposition - дата публикации
# переведем столбец с датой в формат даты без времени, т.к. время не указано
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format = '%Y-%m-%d')
```

```python
# balcony - количество балконов
# видим пустые значения
df['balcony'].unique()
```

```python
df['balcony'].value_counts() # необходимо изменить тип данных
```

```python
# заменим пустые значения на 0 и изменим тип данных
df['balcony'] = df['balcony'].fillna(0).astype('int')
```

```python
# ceiling_height - высота потолков
df['ceiling_height'].value_counts()
```

```python
df[df['ceiling_height']>4] # посмотрим, где высота потолков больше 4 метров
```

```python
# посмотрим living_area жилую площадь
df['living_area'].sort_values() 
# заменим пропущенные значения на среднее mean()
```

```python
dm_mean = df['living_area']/df['total_area'] # разделили жилую площадь на общую
dm_mean = dm_mean.mean()                     # создали переменную, в ней сохранили условное среднее 
dm_mean                                      # вывели значение переменной
```

```python
#spisok_living_area = df.index[df['living_area'].isnull()] # собрали список индексов с пропущенными значениями
```

```python
# заменим отсутствующие значения living_area на : total_area * dm_mean (наше условное среднее)
#for i in spisok_living_area: # пробегаем по нашему списку индексов с пропущенными значениями
#    df.loc[i, 'living_area'] = df.loc[i, 'total_area']*dm_mean 
    # берем каждый индекс из списка и подставляем через loc в нашем df
    # присваем отсутствующему значению = значение общей площади * среднее (total_area * dm_mean)
```

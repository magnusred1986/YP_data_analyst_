#### ТОП 3 книги по средней оценке

```python
query = ''' 

SELECT books.title,
       COUNT (distinct review_id) as review_cnt,
       AVG(rating) AS avg_rating
FROM books
LEFT JOIN reviews ON reviews.book_id = books.book_id
LEFT JOIN ratings ON ratings.book_id = books.book_id

GROUP BY books.book_id
ORDER BY AVG(ratings.rating) DESC
LIMIT 3;

'''

reviews_and_avg_rating_top_3_avg = pd.io.sql.read_sql(query, con = engine)
reviews_and_avg_rating_top_3_avg
```

### Определим издательство, которое выпустило наибольшее число книг `толще 50 страниц` — так вы исключите из анализа брошюры

<a name="stage_1"></a>

[Наверх к оглавлению](#contents)

```python
query = ''' 

SELECT publishers.publisher,

        COUNT(num_pages) as count_books
        
FROM books
INNER JOIN publishers on publishers.publisher_id = books.publisher_id
WHERE books.num_pages > 50
GROUP BY publishers.publisher
ORDER BY count_books DESC 
LIMIT 3

'''

thicker_publishers = pd.io.sql.read_sql(query, con = engine)
thicker_publishers
```

Выведено ТОП-3 издательства выпустивших наибольшее число книг, в которых число страниц > 50  
`Penguin Books` на первом месте  

### Определим автора с самой высокой средней оценкой книг — учитывайте только `книги с 50 и более оценками`  

<a name="stage_1"></a>

[Наверх к оглавлению](#contents)

```python
query = ''' 

SELECT authors.author,
        AVG(ratings.rating) as mean
FROM books
INNER JOIN authors on authors.author_id = books.author_id
INNER JOIN ratings on ratings.book_id = books.book_id
WHERE books.book_id IN (SELECT book_id
                        FROM ratings
                        GROUP BY book_id
                        HAVING COUNT(rating) >= 50)
GROUP BY authors.author
ORDER BY mean DESC
LIMIT 1 


'''    
best_authors = pd.io.sql.read_sql(query, con = engine)
best_authors
```

Автором с самой высокой средней оценкой (среди книг с 50 и более оценками) , является - `J.K. Rowling/Mary GrandPré`

### Посчитаем `среднее количество обзоров` от пользователей, которые поставили `больше 50 оценок`  

<a name="stage_1"></a>

[Наверх к оглавлению](#contents)

```python
query = ''' 


SELECT SUM(group_table.count_rev)/COUNT(group_table.count_rev) as mean
        FROM (SELECT reviews.username, 
        COUNT(reviews.username) AS count_rev
FROM reviews
WHERE reviews.username IN (SELECT ratings.username
                            FROM ratings
                            GROUP BY ratings.username
                            HAVING COUNT(ratings.rating_id) > 50)
                  
                            GROUP BY username) AS group_table 
                            

'''         
review_avg = pd.io.sql.read_sql(query, con = engine)
review_avg
```

Среднее количество обзоров от пользователей `24.3`

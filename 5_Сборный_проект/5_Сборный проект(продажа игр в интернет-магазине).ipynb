Напишем функцию для определения корреляции

```python
def corr_info(x):
    if x <= 0.5:
        return 'корреляция слабая'
    if 0.5 < x <= 0.7:
        return 'корреляция средняя'
    if 0.7 < x <= 0.9:
        return 'корреляция высокая'
    if x > 0.9:
        return 'корреляция очень высокая'
```

`Entertainment Software Rating Board (ESRB)` — негосударственная организация, основное направление деятельности — принятие и определение рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде. ESRB была основана в 1994 году ассоциацией Interactive Digital Software Association (современное название — «Entertainment Software Association»). К началу 2003 года рейтинги ESRB получили более 8000 продуктов от 350 издателей.  

`«M» («Mature») — «Для взрослых»`: Материалы игры не подходят для подростков младше 17 лет.  
`«E» («Everyone») — «Для всех»`: Содержание вполне подходит для возрастной категории от 6 лет.  
`«E10+» («Everyone 10 and older»)` — «Для всех от 10 лет и старше»: Проекты с данным рейтингом могут содержать или немного больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови. Рейтинг был принят ESRB 2 марта 2005 года.  
`«T» («Teen»)` — «Подросткам»: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.  
`«RP» («Rating Pending»)` — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

`H0:(нулевая гипотеза)` Средние пользовательские рейтинги платформ Xbox One и PC одинаковые  
`H1:(альтернативная гипотеза)` Средние пользовательские рейтинги платформ Xbox One и PC отличаются

```python
alpha = 0.05
```

```python
XOne = df[(df['platform'] == 'XOne') & (df['user_score'] > 0)]['user_score']
PC = df[(df['platform'] == 'PC') & (df['user_score'] > 0)]['user_score']
```

```python
results = st.ttest_ind(XOne, PC)
```

```python
XOne_mean = XOne.mean()
PC_mean = PC.mean()
difference_XOne_PC = abs(XOne_mean-PC_mean)
```

```python
print('t-значение:', results.statistic)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print('Среднее по XOne ', XOne_mean.round(2))
print('Среднее по PC ', PC_mean.round(2))
print('Разница средних ', difference_XOne_PC.round(2))
```

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

`H0:(нулевая гипотеза)` Средние пользовательские рейтинги жанров Action и Sports одинаковые  
`H1:(альтернативная гипотеза)` Средние пользовательские рейтинги жанров Action и Sports разные  

```python
alpha = 0.05
```

```python
action_user_score = df[(df['genre'] == 'Action') & (df['user_score'] > 0)]['user_score']
sports_user_score = df[(df['genre'] == 'Sports') & (df['user_score'] > 0)]['user_score']
```

```python
results = st.ttest_ind(action_user_score, sports_user_score)
```

```python
action_user_score_mean = action_user_score.mean()
sports_user_score_mean = sports_user_score.mean()
difference_action_sports = abs(action_user_score_mean-sports_user_score_mean)
```

```python
print('t-значение:', results.statistic)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print('Среднее по Action ', action_user_score_mean.round(2))
print('Среднее по Sports ', sports_user_score_mean.round(2))
print('Разница средних ', difference_action_sports.round(2))
```

На данном этапе, верное утверждение **Средние пользовательские рейтинги жанров Action и Sports одинаковые**

```python
def hypothesis(df, colum1 = ('genre'), fil_1_colum1 = 'Action', fil_2_colum1 = 'Sports', colum2 = 'user_score'):
    '''
    Функция - проверки гипотез
    df = имя датафрейма
    colum1 = ('genre') или ('platform') столбец фильтрации 1
    fil_1_colum1 = 'Action' фильтр первого критерия по colum1
    fil_2_colum1 = 'Sports' фильтр второго критерия по colum1
    colum2 = 'user_score' столбец фильтрации 2
    '''
    alpha = 0.05
    name1 = fil_1_colum1
    name2 = fil_2_colum1
    fil_1_colum1 = df[(df[colum1] == fil_1_colum1) & (df[colum2] > 0)][colum2]
    fil_2_colum1 = df[(df[colum1] == fil_2_colum1) & (df[colum2] > 0)][colum2]
    results = st.ttest_ind(fil_1_colum1, fil_2_colum1)
    fil_1_colum1_mean = fil_1_colum1.mean()
    fil_2_colum1_mean = fil_2_colum1.mean()
    difference = abs(fil_1_colum1_mean-fil_2_colum1_mean)
    print('t-значение:', results.statistic)
    print('p-значение:', results.pvalue)
    if (results.pvalue < alpha):
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")
    print('Среднее по '+ " "+ name1, fil_1_colum1_mean)
    print('Среднее по '+ " "+ name2, fil_2_colum1_mean)
    print('Разница средних ', difference)
```

```python
hypothesis(df, colum1 = ('platform'), fil_1_colum1 = 'XOne', fil_2_colum1 = 'PC', colum2 = 'user_score')
```

```python
hypothesis(df, colum1 = ('genre'), fil_1_colum1 = 'Action', fil_2_colum1 = 'Sports', colum2 = 'user_score')
```
